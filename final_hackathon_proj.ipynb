```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
import os
import cv2
import matplotlib.pyplot as plt


# takes data set and loads it
class MRIDataset(Dataset):
   def __init__(self, root_dir):
       data_path = os.path.join(root_dir, "Training", "Tr-gl_0375")
       self.files = [os.path.join(data_path, f) for f in os.listdir(data_path) if f.endswith(('.jpg', '.png'))]
                    
   def __len__(self):
       return len(self.files)

   #takes uploaded files,resize,and grey scaled it so that it would be easier to get unbiased data
   def __getitem__(self, idx):
       img_path = self.files[idx]
       image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
       image = cv2.resize(image, (128, 128)) / 255.0 
       image = torch.tensor(image, dtype=torch.float32).unsqueeze(0) * 2 - 1
       return image


# generator - generates the 'fake' mri images, transpose multiple layers until 128x128 to get more defined images
class Generator(nn.Module):
   def __init__(self, noise_dim=100):
       super(Generator, self).__init__()
       self.fc = nn.Linear(noise_dim, 256 * 4 * 4)
       self.model = nn.Sequential(
           nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1),
           nn.BatchNorm2d(128),
           nn.ReLU(inplace=True),

           nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1),
           nn.BatchNorm2d(64),
           nn.ReLU(inplace=True),

           nn.ConvTranspose2d(64, 32, kernel_size=4, stride=2, padding=1),
           nn.BatchNorm2d(32),
           nn.ReLU(inplace=True),

           nn.ConvTranspose2d(32, 16, kernel_size=4, stride=2, padding=1),
           nn.BatchNorm2d(16),
           nn.ReLU(inplace=True),
           nn.ConvTranspose2d(16, 1, kernel_size=4, stride=2, padding=1),
           nn.Tanh()
       )
   def forward(self, x):
       x = x.view(x.size(0), -1)
       x = self.fc(x).view(x.size(0), 256, 4, 4)
       return self.model(x)


# discriminator - will use this code to determine the 'realness' of the generated MRIs.
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.model = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=4, stride=2, padding=1),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Dropout(0.3), 

            nn.Conv2d(16, 32, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(32),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Dropout(0.3), 

            nn.Conv2d(32, 64, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(128, 1, kernel_size=4, stride=1, padding=0),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.model(x)


#parameters for imaging
z_dim = 100
lr = 0.0001
batch_size = 64
runs = 1000

#initializing the model
generator = Generator(z_dim)
discriminator = Discriminator()
criterion = nn.BCELoss()

optimizer_g = optim.Adam(generator.parameters(), lr=lr)
optimizer_d = optim.Adam(discriminator.parameters(), lr=lr * 0.4)  # Reduce by half


train_dataset = MRIDataset(root_dir="brain_mri_images")
train_load = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

for i in range(runs):
    for real_scans in train_load:
        batch_size = real_scans.size(0)

        z = torch.randn(batch_size, z_dim, 1, 1)
        fake_scans = generator(z)

        actual_preds = discriminator(real_scans)
        fake_preds = discriminator(fake_scans.detach())

        d_loss = criterion(actual_preds, torch.ones_like(actual_preds) * 0.9) + \
                 criterion(fake_preds, torch.zeros_like(fake_preds))
        
        optimizer_d.zero_grad()
        d_loss.backward()
        optimizer_d.step()

        fake_preds_gen = discriminator(fake_scans)
        g_loss = criterion(fake_preds_gen, torch.ones_like(fake_preds_gen))
        
        optimizer_g.zero_grad()
        g_loss.backward()
        optimizer_g.step()
    print("Epoche:", i+1, "/", runs, ", Discriminator Loss:", d_loss.item(), ", Generator Loss:", g_loss.item())
      
   # visualize generated images periodically
    if (i + 1) % 100 == 0:
       z = torch.randn(1, z_dim, 1, 1)
       fake_image = generator(z).squeeze().detach().numpy()
       plt.imshow(fake_image, cmap='gray')
       plt.axis('off')
       plt.show()
```
